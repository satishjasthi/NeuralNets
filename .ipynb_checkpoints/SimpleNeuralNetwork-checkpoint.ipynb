#### An example to show how tf.multiply does element wise multiplication 

```python
with tf.Graph().as_default() as graph:
    
    #inputs from input layer,which is just 3 values
    x = tf.convert_to_tensor(np.random.normal(size = (3,1))) 
    #or
    x = tf.random_normal((3,1),name="x")
    
    '''
    weights are of shape (4,3) because there are three features for each 
    sample/example and each feature needs to connect with 4 neurons in hidden
    layer_1
    '''
    w1 = tf.convert_to_tensor(np.random.normal(size=(4,3)))
    #or
    w1 = tf.random_normal((4,3),name="w1")
    
    #hidden layer, where we get both z = wT*x and a = sigmoid(z)
    hl1 = tf.matmul(w1,x,name="hl1")
    b1 = tf.ones((w1.shape[0],1),name="b1")
    a1 = tf.sigmoid(hl1+b1,name="a1")
    
    '''
    weights between hidden layer and output layer is of shape(2,4), since there
    are 4 inputs and two outputs
    '''
    w2 = tf.random_normal((2,4),name="w2")
    
    #output layer
    ol = tf.matmul(w2,a1,name="ol")
    b2 = tf.ones((w2.shape[0],1),name="b2")
    a2 = tf.sigmoid(ol+b2,name="output")
    
    with tf.Session(graph=graph) as sess:
        writer = tf.summary.FileWriter("./graphs",graph=sess.graph)
        print sess.run(a2)
      
        writer.close()
os.system("tensorboard --logdir="+"./graphs")
        
```

## Implementing NN for multiple samples/examples

```python
with tf.Graph().as_default() as graph:
    #hidden layer and its activation
    x = tf.random_normal((3,100),name="x")
    w1 = tf.random_normal((4,3),name="w1")
    hl1 = tf.matmul(w1,x,name="hl1")
    b1 = tf.random_normal((w1.get_shape().as_list()[0],1),name="b1") #initializing bais with random values instead of ones or zeros
    a1 = tf.sigmoid(hl1+b1,name="a1")
    
    #output layer and its activation
    w2 = tf.random_normal((2,w1.get_shape().as_list()[0]),name = "w2")
    ol = tf.matmul(w2,a1,name="ol")
    b2 = tf.random_normal((w2.get_shape().as_list()[0],1),name="b2")
    a2 = tf.sigmoid(ol+b2,name="a2")
    
    
    
    with tf.Session(graph=graph) as sess:
        writer = tf.summary.FileWriter("./graphs",graph=sess.graph)
        print sess.run(a1)
        writer.close()
os.system("tensorboard --logdir="+"./graphs")
```
